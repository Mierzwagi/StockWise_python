```python
import pandas as pd
import numpy as np
import json

tabela = pd.read_excel('./data/PATRIMONIO 12 2023.xlsx')

display(tabela)
```

```python
display(tabela.info())

tabela['Nº inventário'].fillna(0, inplace=True)
tabela['Nº inventário'] = tabela['Nº inventário'].astype(int)
tabela['Incorporação em'] = tabela['Incorporação em'].astype('string')
""" tabela['Incorporação em'] = tabela['Incorporação em'].dt.strftime('%y-%m-%d') """

display(tabela.info())
```

```python
resultado = {}

for _, row in tabela.iterrows():
    resultado[row['Denominação']] = {
        'Nº inventário': row['Nº inventário'],
        'localização': row['Localização'],
        'Incorporação em': row['Incorporação em']
    }

resultado_json = json.dumps(resultado, indent=2, ensure_ascii=False)

print(resultado_json)
```

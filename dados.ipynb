```python
import pandas as pd
import numpy as np

tabela = pd.read_csv("PATRIMONIO 12 2023.csv", encoding='windows-1252', delimiter=';')

display(tabela)


```

```python
display(tabela.info())

tabela['Nº inventário'].fillna(0, inplace=True)

tabela['Nº inventário'] = tabela['Nº inventário'].astype(int)

display(tabela.info())
```

```python
tabela_sala21080001 = tabela[tabela['Localização'] == 21080001]
display(tabela_sala21080001)
display(tabela_sala21080001["Nº inventário"].value_counts())

tabela_sala21080013 = tabela[tabela['Localização'] == 21080013]
display(tabela_sala21080013)
display(tabela_sala21080013["Nº inventário"].value_counts())
```

```python
tabela_sala21080013['Sala'] = np.where(tabela_sala21080013['Localização'] == 21080013, 'SALA 13', 'OUTRA')
display(tabela_sala21080013)
```

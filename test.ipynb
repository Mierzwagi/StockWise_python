```python
import pandas as pd
import numpy as np
import json
```

```python
df_items = pd.read_csv("PATRIMONIO 12 2023.csv", encoding="windows-1252", delimiter=";")
df_salas = pd.read_csv(
    "CÓDIGO AMBIENTE SIG 2023(Planilha1).csv", encoding="windows-1252", delimiter=";"
)
```

```python
df_items['Id'] = df_items['Id'].fillna(0)
```

```python
df_items['Id'] = df_items['Id'].astype(int)
```

```python
df_salas.columns
```

```python
df_resultado = pd.merge(df_items, df_salas, on='Localizacao', how='inner')

df_resultado.tail()
```

```python
groupLocalizacao = df_resultado.groupby('Localizacao')

type(groupLocalizacao)
```

```python
import json


# Inicializa o dicionário resultante
result = {}

# Itera sobre cada grupo de Localização
for localizacao, grupo in groupLocalizacao:
    items = grupo[["Id", "Denominacao", "Incorporacao em"]].to_dict("records")
    quantidade_inventario = int(grupo["Id"].count())
    sala = ", ".join(grupo["Ambiente"].unique())  # Junta todos os valores únicos de "Ambiente"

    # Prepara o dicionário resultante
    result[localizacao] = {
        "quantidade de Nº inventário": quantidade_inventario,
        "Sala": sala,
        "items": [
            {"id": item["Id"], "denominacao": item["Denominacao"], "dataDeIncorporacao": item["Incorporacao em"]} for item in items
        ],
    }

# Converte o dicionário resultante para JSON
resultado_json = json.dumps(result, indent=2, ensure_ascii=False)

# Exibe o JSON resultante
print(resultado_json)
```
